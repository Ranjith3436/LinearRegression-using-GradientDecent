```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
plt.rcParams['figure.figsize'] = (12.0, 9.0)
data=pd.read_csv("C:/Users/Lenovo/OneDrive/Desktop/Dataset.csv")
# data
# sns.regplot(x=data['Input'],y=data['Output'])
X=data.loc[:,("Input")]
y=data.loc[:,("Output")]
plt.scatter(X, y)
plt.show()
```

```python
m=1
c=0
iteration=1000
error=0
cost=[]
itr=[]
L=0.0001
n=100

plt.rcParams['figure.figsize'] = (16, 8)
for i in range(iteration):
    ypred=m*X+c
    error=(0.5/n)*sum([val**2 for val in (y-ypred)])
    D_m = (-2/n) * sum(X * (y- ypred))  # Derivative wrt m
    D_c = (-2/n) * sum(y- ypred)  # Derivative wrt c
    m = m - L * D_m  # Update m
    c = c - L * D_c  # Update 
    cost.append(error)
    itr.append(i)
    
y_1=m*X+c   
    
print(m,c,error)
e=((y-ypred)**2)**0.5
plt.plot(itr,cost)
# plt.plot(itr[400:],cost[400:])
# plt.scatter(X,e)

# plt.plot(X,y_1,color='red')
# plt.scatter(X, y)

# plt.plot([min(X), max(X)], [min(ypred), max(ypred)], color='red')  # regression line
# plt.show()
```
